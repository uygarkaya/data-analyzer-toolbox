### Data Processing Notebook
Clean, Transform, and Prepare Your Dataset for Analysis

```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
pd.set_option("display.width", 120)
```

### Load Dataset

```python
df = pd.read_csv('../data/example/bmw.csv')
print(f"Dataset Loaded: {df.shape[0]:,} rows × {df.shape[1]} columns\n")
df.head()
```

### Quick Overview

```python
print("Data Types")
print(df.dtypes.to_string())
print()
print("Missing Values Per Column")
nulls = df.isnull().sum()
nulls_pct = (nulls / len(df) * 100).round(2)
missing_info = pd.DataFrame({"Missing": nulls, "% of Total": nulls_pct})
print(missing_info[missing_info["Missing"] > 0].to_string() if nulls.sum() > 0 else "No Missing Values Found!")
print()
print(f"Duplicate Rows: {df.duplicated().sum():,}")
print(f"Total Rows:     {len(df):,}")
```


### 1. Handle Missing Values

```python
column = "Month"  # NOTE: please change to column name
strategy = "mean" # NOTE: options -> mean, median, mode, constant, drop
constant_value = 0 # NOTE: please change if strategy is "constant"

before_nulls = df[column].isnull().sum()
print(f"Before: '{column}' has {before_nulls} Missing Value(s)")

if before_nulls == 0:
    print("→ No Missing Values — Nothing To Do!")
else:
    if strategy == "mean":
        df[column] = df[column].fillna(df[column].mean())
    elif strategy == "median":
        df[column] = df[column].fillna(df[column].median())
    elif strategy == "mode":
        df[column] = df[column].fillna(df[column].mode()[0])
    elif strategy == "constant":
        df[column] = df[column].fillna(constant_value)
    elif strategy == "drop":
        df = df.dropna(subset=[column])

    after_nulls = df[column].isnull().sum()
    print(f"After:  '{column}' has {after_nulls} Missing Value(s)")
    print(f"{before_nulls - after_nulls} Null(s) Handled Using '{strategy}'")

print(f"\nDataset Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
```

### Remove Duplicate Rows

```python
subset_cols = None # NOTE: specify columns to check for duplicates (e.g., ["col1", "col2"]) or set to None to check all columns
keep = "first" # NOTE: "first" (keep first occurrence), "last" (keep last occurrence), False (drop all duplicates)

before = len(df)
df = df.drop_duplicates(subset=subset_cols, keep=keep)
removed = before - len(df)

print(f"Before: {before:,} rows")
print(f"After:  {len(df):,} rows")
print(f"Removed {removed:,} duplicate row(s)")
```


### Rename / Drop Columns

3a. Rename a Column

```python
old_name = "Month"
new_name = "Month_Num"

if old_name not in df.columns:
    print(f"Column '{old_name}' Not Found! Available: {list(df.columns)}")
elif new_name in df.columns:
    print(f"Column '{new_name}' Already Exists!")
else:
    df = df.rename(columns={old_name: new_name})
    print(f"'{old_name}' → '{new_name}'")

print(f"Columns: {list(df.columns)}")
```

3b. Drop a Column

```python
drop_column = "Region"

if drop_column not in df.columns:
    print(f"Column '{drop_column}' Not Found! Available: {list(df.columns)}")
else:
    df = df.drop(columns=[drop_column])
    print(f"Column '{drop_column}' Dropped! {len(df.columns)} Column(s) Remaining!")

print(f"Columns: {list(df.columns)}")
```

### Final Dataset Preview

```python
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Duplicate rows: {df.duplicated().sum()}")
print()
df.head(10)
```

### Export Processed Dataset

```python
output_path = "../data/processed/bmw.csv"
df.to_csv(output_path, index=False)
print(f"Saved to '{output_path}' ({df.shape[0]:,} rows × {df.shape[1]} columns)")
```
