### Data Overview
View Dataset Summary, Data Types, Missing Values, and More!

```python
import pandas as pd
```

### Load Dataset
Replace the path below with your CSV file path.

```python
df = pd.read_csv('../data/example/bmw.csv')
print(f'Dataset loaded: {df.shape[0]:,} rows × {df.shape[1]} columns')
```

### Summary Statistics

```python
total_rows = len(df)
total_cols = len(df.columns)
total_missing = int(df.isnull().sum().sum())
duplicates = int(df.duplicated().sum())
numeric_cols = len(df.select_dtypes(include='number').columns)
cat_cols = len(df.select_dtypes(include=['object', 'category']).columns)
dt_cols = len(df.select_dtypes(include=['datetime', 'datetimetz']).columns)
memory_bytes = df.memory_usage(deep=True).sum()
memory_str = (f'{memory_bytes / 1024:.1f} KB' 
              if memory_bytes < 1_048_576 
              else f'{memory_bytes / 1_048_576:.2f} MB')

summary = pd.DataFrame({
    'Metric': [
        'Total Rows', 'Total Columns', 'Missing Values',
        'Duplicate Rows', 'Numeric Columns',
        'Categorical Columns', 'Datetime Columns', 'Memory Usage'
    ],
    'Value': [
        f'{total_rows:,}', f'{total_cols:,}', f'{total_missing:,}',
        f'{duplicates:,}', str(numeric_cols),
        str(cat_cols), str(dt_cols), memory_str
    ]
})
summary
```

### Column Details

```python
col_rows = []
for col in df.columns:
    null_count = int(df[col].isnull().sum())
    col_rows.append({
        'Column': col,
        'Data Type': str(df[col].dtype),
        'Non-Null': total_rows - null_count,
        'Null Count': null_count,
        'Null %': f'{null_count / total_rows * 100:.1f}%',
        'Unique Values': int(df[col].nunique(dropna=False)),
        'Sample Value': str(df[col].dropna().iloc[0]) if not df[col].dropna().empty else 'N/A'
    })

col_details = pd.DataFrame(col_rows)
col_details
```

### Numeric Summary Statistics

```python
num_df = df.select_dtypes(include='number')
if not num_df.empty:
    display(num_df.describe().round(4))
else:
    print('No numeric columns found.')
```
